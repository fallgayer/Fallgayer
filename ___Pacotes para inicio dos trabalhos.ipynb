```python
import seaborn as sns
sns.set_theme(style="whitegrid")

# Load the brain networks dataset, select subset, and collapse the multi-index
df = sns.load_dataset("brain_networks", header=[0, 1, 2], index_col=0)

used_networks = [1, 5, 6, 7, 8, 12, 13, 17]
used_columns = (df.columns
                  .get_level_values("network")
                  .astype(int)
                  .isin(used_networks))
df = df.loc[:, used_columns]

df.columns = df.columns.map("-".join)

# Compute a correlation matrix and convert to long-form
corr_mat = df.corr().stack().reset_index(name="correlation")

# Draw each cell as a scatter point with varying size and color
g = sns.relplot(
    data=corr_mat,
    x="level_0", y="level_1", hue="correlation", size="correlation",
    palette="vlag", hue_norm=(-1, 1), edgecolor=".7",
    height=10, sizes=(50, 250), size_norm=(-.2, .8),
)

# Tweak the figure to finalize
g.set(xlabel="", ylabel="", aspect="equal")
g.despine(left=True, bottom=True)
g.ax.margins(.02)
for label in g.ax.get_xticklabels():
    label.set_rotation(90)
```

```python
# Ajustando para ver uma função quadrática
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
```

```python
# Criar a ordem do polinômio (2,3,4,...)
poly_features = PolynomialFeatures(degree=1)
X_poly = poly_features.fit_transform(X)
```

```python
# Criar e ajustar o modelo de regressão polinomial
model = LinearRegression()
model.fit(X_poly, y)
```

```python
# Obter os coeficientes da regressão
intercepto = model.intercept_
coeficientes = model.coef_
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

# Definir o grau do polinômio
degree = 5  # <<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<  Escolher o melhor grau <<<<<<<<<<<<<<<<<<

# Criar características polinomiais
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)

# Criar e ajustar o modelo de regressão polinomial
model = LinearRegression()
model.fit(X_poly, y)

# Obter os coeficientes da regressão
intercepto = model.intercept_
coeficientes = model.coef_

# Construir a função de regressão como uma string
def construir_funcao_polinomial(intercepto, coeficientes):
    termos = []
    for i, coef in enumerate(coeficientes):
        if coef == 0:
            continue
        potencia = i
        if potencia == 0:
            termos.append(f"{coef:.2f}")
        elif potencia == 1:
            termos.append(f"{coef:.2f}x")
        else:
            termos.append(f"{coef:.2f}x^{potencia}")

    funcao = " + ".join(termos)
    return f"f(x) = {funcao}"

# Obter a função no formato desejado
funcao = construir_funcao_polinomial(intercepto, coeficientes)
print(funcao)

# Visualizar os resultados
plt.scatter(X, y, color='blue', label='Dados')

# Gerar valores preditos para a linha de regressão
X_range = np.linspace(min(X), max(X), 100).reshape(-1, 1)
X_range_poly = poly_features.transform(X_range)
y_pred = model.predict(X_range_poly)

plt.plot(X_range, y_pred, color='red', linewidth=2, label='Linha de Regressão Polinomial')

# Configurar os limites dos eixos para dar zoom no ponto (0,0)
plt.xlim(min(X), max(X))
plt.ylim(min(y), max(y))

# Adicionar linhas de referência para o ponto (0,0)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)

plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Regressão Polinomial (grau {degree})')
plt.legend()
plt.grid(True)
plt.show()
```
