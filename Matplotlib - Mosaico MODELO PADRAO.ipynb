```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Carregar conjunto de dados "tips" do seaborn
tips = sns.load_dataset("tips")
```

```python
# Criar uma figura com um mosaico de 2x2
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Grfico de barras (barplot)
sns.barplot(x='day', y='total_bill', data=tips, ax=axs[0, 0], ci='sd', palette='viridis')
axs[0, 0].set_title('Total Bill per Day')
axs[0, 0].set_xlabel('Day')
axs[0, 0].set_ylabel('Total Bill')

# Grfico de linha (lineplot)
sns.lineplot(x='size', y='total_bill', data=tips, ax=axs[0, 1], ci='sd', marker='o', linewidth=2.5)
axs[0, 1].set_title('Total Bill by Size of Party')
axs[0, 1].set_xlabel('Size of Party')
axs[0, 1].set_ylabel('Total Bill')

# Grfico de disperso (scatterplot)
sns.scatterplot(x='total_bill', y='tip', data=tips, ax=axs[1, 0], hue='day', palette='deep', s=100, edgecolor='w', linewidth=1.5)
axs[1, 0].set_title('Tip vs Total Bill')
axs[1, 0].set_xlabel('Total Bill')
axs[1, 0].set_ylabel('Tip')

# Grfico de violino (violinplot)
sns.violinplot(x='day', y='total_bill', data=tips, ax=axs[1, 1], inner='quartile', palette='pastel')
axs[1, 1].set_title('Distribution of Total Bill by Day')
axs[1, 1].set_xlabel('Day')
axs[1, 1].set_ylabel('Total Bill')

# Ajustar layout
plt.tight_layout()
plt.show()
```
