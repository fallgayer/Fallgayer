```python
import pandas as pd

# Criar um DataFrame simples
data = {
    'Nome': ['Alice', 'Bob', 'Carol', 'David'],
    'Idade': [25, 30, 22, 18],
    'Salário': [3500.50, 4200.75, 2800.25, 3200.00]
}

df = pd.DataFrame(data)

# Exemplo 1: Exibindo uma variável dentro de uma string
mensagem1 = f"Olá, {df['Nome'][0]}!"
print(mensagem1)  # Saída: Olá, Alice!

# Exemplo 2: Realizando operações dentro da f-string
mensagem2 = f"O dobro do salário de {df['Nome'][1]} é {df['Salário'][1] * 2}."
print(mensagem2)  # Saída: O dobro do salário de Bob é 8401.5.

# Exemplo 3: Formatando um número com diferentes precisões
mensagem3 = f"O salário de {df['Nome'][2]} é R$ {df['Salário'][2]:.2f}."
print(mensagem3)  # Saída: O salário de Carol é R$ 2800.25.

# Exemplo 4: Usando expressões condicionais dentro da f-string
mensagem4 = f"{df['Nome'][3]} é {'maior' if df['Idade'][3] >= 18 else 'menor'} de idade."
print(mensagem4)  # Saída: David é maior de idade.

# Exemplo 5: Utilizando formatação de string para preencher espaços com um número específico de caracteres
mensagem5 = f"Olá, {df['Nome'][0]:^10}!"
print(mensagem5)  # Saída: Olá,   Alice   !
```

```python
import pandas as pd

# Criar um DataFrame com datas
data = {
    'Data': ['2022-01-15', '2023-05-20', '2024-10-25', '2025-12-31']
}

df = pd.DataFrame(data)

# Converter a coluna 'Data' para o tipo datetime
df['Data'] = pd.to_datetime(df['Data'])

# Adicionar colunas para o primeiro e o último dia do mês
df['Primeiro Dia do Mês'] = df['Data'].dt.to_period('M').dt.to_timestamp()
df['Último Dia do Mês'] = df['Data'].dt.to_period('M').dt.to_timestamp(freq='M') + pd.DateOffset(days=-1)

# Aplicar diversos formatos de string para as datas
df['Formato1'] = df['Data'].dt.strftime("%d/%m/%Y")  # Formato: Dia/Mês/Ano
df['Formato2'] = df['Data'].dt.strftime("%B %d, %Y")  # Formato: Nome do Mês Dia, Ano
df['Formato3'] = df['Data'].dt.strftime("%Y-%m-%d")  # Formato: Ano-Mês-Dia
df['Formato4'] = df['Data'].dt.strftime("%A, %d de %B de %Y")  # Formato: Dia da Semana, Dia de Mês de Ano

# Exibir o DataFrame com os diferentes formatos de datas, primeiro e último dia do mês
print(df)
```

```python
# Um estilo visual com visão dos quartis coloridos

import pandas as pd
import numpy as np

# Criando um DataFrame de exemplo com valores aleatórios
np.random.seed(0)
data = np.random.randn(10, 2)
df = pd.DataFrame(data, columns=['Coluna 1', 'Coluna 2'])

# Calculando os quartis
q1 = df.quantile(0.25)
q2 = df.quantile(0.50)
q3 = df.quantile(0.75)

df.sort_values(by='Coluna 1', inplace=True)

# Função para definir as cores dos quartis
def quartile_colors(s):
    colors = []
    for val in s:
        if val <= q1:
            colors.append('#FF5733')  # Cor quente para o primeiro quartil
        elif val <= q2:
            colors.append('#FFC300')  # Cor morna para o segundo quartil
        elif val <= q3:
            colors.append('#85C1E9')  # Cor fria para o terceiro quartil
        else:
            colors.append('#4B9CD3')  # Cor mais fria para o quarto quartil
    return colors

# Aplicando o gradiente de fundo com base nos quartis
styled_df = df.style.background_gradient(cmap='coolwarm', subset=df.columns)

# Exibindo o DataFrame com o gradiente de fundo aplicado
styled_df
```
