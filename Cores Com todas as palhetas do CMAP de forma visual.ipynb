```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib import cm

# Obtendo a lista de nomes de todos os mapas de cores disponíveis
colormaps = plt.colormaps()

# Definindo o número de graduações de cores para cada mapa de cores
num_gradations = 15

# Lista de dicionários para armazenar as cores de cada mapa de cores
colors_list = []

# Criando um DataFrame com valores para cada mapa de cores
for cmap in colormaps:
    cmap_obj = cm.get_cmap(cmap)
    colors = [cmap_obj(x) for x in np.linspace(0, 1, num_gradations)]
    hex_colors = [mcolors.to_hex(color) for color in colors]
    colors_dict = {'Mapa de Cores': cmap, 'Cores': hex_colors}
    colors_list.append(colors_dict)

# Criando um DataFrame a partir da lista de dicionários
color_df = pd.DataFrame(colors_list)

# Expandindo a coluna 'Cores' em múltiplas colunas
color_df_expanded = pd.DataFrame(color_df['Cores'].tolist(), index=color_df['Mapa de Cores'])

# Estilizando o DataFrame para exibir as cores
styled_color_df = color_df_expanded.style.applymap(lambda x: f'background-color: {x}')

# Exibindo o DataFrame estilizado
styled_color_df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib import cm
from openpyxl import Workbook
import weasyprint

# Obtendo a lista de nomes de todos os mapas de cores disponíveis
colormaps = plt.colormaps()

# Definindo o número de graduações de cores para cada mapa de cores
num_gradations = 7

# Criando um DataFrame vazio para armazenar as cores
color_df = pd.DataFrame()

# Criando um DataFrame com valores para cada mapa de cores
for cmap in colormaps:
    colors = [cm.get_cmap(cmap)(x) for x in np.linspace(0, 1, num_gradations)]
    hex_colors = [mcolors.to_hex(color) for color in colors]
    color_df[cmap] = hex_colors

# Transpondo o DataFrame para que cada linha represente um mapa de cores
color_df = color_df.T

# Estilizando o DataFrame para exibir as cores
styled_color_df = color_df.style.applymap(lambda x: f'background-color: {x}')

# Exportando para HTML
styled_color_df.to_html('cores.html', escape=False, index=False)

# Exportando para PDF
wb = Workbook()
ws = wb.active
ws.append(styled_color_df)
wb.save('cores.xlsx')
weasyprint.HTML('cores.html').write_pdf('cores.pdf')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# Obtendo a lista de nomes de todos os mapas de cores disponíveis
colormaps = plt.colormaps()

# Definindo o número de graduações de cores para cada mapa de cores
num_gradations = 7

# Criando um DataFrame vazio
color_df = pd.DataFrame()

# Criando um DataFrame com valores para cada mapa de cores
for cmap in colormaps:
    colors = [cm.get_cmap(cmap)(x) for x in np.linspace(0, 1, num_gradations)]
    color_df[cmap] = colors

# Exibindo o DataFrame com as cores para cada mapa de cores
color_df.style.applymap(lambda x: 'background-color: rgba({}, {}, {}, {})'.format(int(x[0]*255), int(x[1]*255), int(x[2]*255), x[3]))

```
