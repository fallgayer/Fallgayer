```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm  # Importar norm para a curva de distribuição normal

# Gerar dados aleatórios com distribuição normal
np.random.seed(0)
data_normal = np.random.normal(loc=0, scale=1, size=1000)

# Criar DataFrame
df_normal = pd.DataFrame(data_normal, columns=['Valor'])

# Adicionar coluna com baixa skewness
df_normal['Baixa_Skewness'] = df_normal['Valor']

# Visualizar primeiras linhas do DataFrame
print("Exemplo com Baixa Kurtosis e Skewness:")
print(df_normal.head())

# Gerar dados com alta kurtosis e skewness
data_high_kurtosis = np.random.gamma(shape=2, scale=1, size=1000)
data_skewed = np.random.gamma(shape=2, scale=2, size=1000)

# Criar DataFrame
df_high_kurtosis = pd.DataFrame({
    'Alta_Kurtosis': data_high_kurtosis,
    'Alta_Skewness': data_skewed
})

# Visualizar primeiras linhas do DataFrame
print("\nExemplo com Alta Kurtosis e Skewness:")
print(df_high_kurtosis.head())

# Plot para o DataFrame com distribuição normal
plt.figure(figsize=(12, 6))
sns.histplot(df_normal['Valor'], kde=True, stat='density', color='blue', bins=30)

# Adicionar curva de distribuição normal
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, df_normal['Valor'].mean(), df_normal['Valor'].std())
plt.plot(x, p, 'k', linewidth=2)
plt.title('Distribuição Normal (Baixa Kurtosis e Skewness)')
plt.xlabel('Valores')
plt.ylabel('Densidade')
plt.show()

# Plot para o DataFrame com alta kurtosis e skewness
plt.figure(figsize=(12, 6))
sns.histplot(df_high_kurtosis['Alta_Kurtosis'], kde=True, stat='density', color='green', bins=30)

# Adicionar curva de distribuição normal
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, df_high_kurtosis['Alta_Kurtosis'].mean(), df_high_kurtosis['Alta_Kurtosis'].std())
plt.plot(x, p, 'k', linewidth=2)
plt.title('Distribuição com Alta Kurtosis')
plt.xlabel('Valores')
plt.ylabel('Densidade')
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, kstest, normaltest, anderson, probplot

# Gerar dados aleatórios com distribuição normal
np.random.seed(0)
data_normal = np.random.normal(loc=0, scale=1, size=1000)

# Criar DataFrame
df = pd.DataFrame(data_normal, columns=['Valor'])

# Histograma
plt.figure(figsize=(12, 6))
sns.histplot(df['Valor'], kde=True, stat='density', bins=30, color='blue')
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Densidade')
plt.show()

# Q-Q Plot
plt.figure(figsize=(6, 6))
probplot(df['Valor'], dist="norm", plot=plt)
plt.title('Q-Q Plot')
plt.show()

# Teste de Shapiro-Wilk
shapiro_stat, shapiro_p = shapiro(df['Valor'])
print(f"Shapiro-Wilk: estatística={shapiro_stat:.3f}, p-valor={shapiro_p:.3f}")

# Teste de Kolmogorov-Smirnov
ks_stat, ks_p = kstest(df['Valor'], 'norm', args=(df['Valor'].mean(), df['Valor'].std()))
print(f"Kolmogorov-Smirnov: estatística={ks_stat:.3f}, p-valor={ks_p:.3f}")

# Teste de Anderson-Darling
ad_result = anderson(df['Valor'])
print("Anderson-Darling:")
print(f"Estatística: {ad_result.statistic:.3f}")
for i, (crit_val, sig_level) in enumerate(zip(ad_result.critical_values, ad_result.significance_level)):
    print(f"Nível de significância: {sig_level:.3f}%, Valor crítico: {crit_val:.3f}")

# Teste de Jarque-Bera
jb_stat, jb_p = normaltest(df['Valor'])
print(f"Jarque-Bera: estatística={jb_stat:.3f}, p-valor={jb_p:.3f}")
```

3. Teste de Anderson-Darling

4. Teste de Jarque-Bera

O p-valor de 0.000 é significativamente menor que 0.05, sugerindo novamente que os dados não seguem uma distribuição normal.

Conclusão
Com base nos resultados desses testes estatísticos, todos indicam que os dados não são normalmente distribuídos. A hipótese nula de normalidade é consistentemente rejeitada em favor da hipótese alternativa de que os dados não são normais.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, norm

# Carregar dados
penguins = sns.load_dataset("penguins")

# Filtrar por espécie, por exemplo, "Adelie"
adelie_males = penguins[(penguins['species'] == 'Adelie') & (penguins['sex'] == 'Male')]
adelie_both_sexes = penguins[penguins['species'] == 'Adelie']

# Testar normalidade para machos
shapiro_males = shapiro(adelie_males['body_mass_g'].dropna())
print(f"Shapiro-Wilk para machos: estatística={shapiro_males.statistic:.3f}, p-valor={shapiro_males.pvalue:.3f}")

# Testar normalidade para ambos os sexos
shapiro_both = shapiro(adelie_both_sexes['body_mass_g'].dropna())
print(f"Shapiro-Wilk para ambos os sexos: estatística={shapiro_both.statistic:.3f}, p-valor={shapiro_both.pvalue:.3f}")

# Visualização
plt.figure(figsize=(12, 6))
sns.histplot(adelie_both_sexes, x='body_mass_g', hue='sex', kde=True, bins=30)
plt.title('Distribuição de Peso dos Pinguins Adelie por Sexo')
plt.xlabel('Peso Corporal (g)')
plt.ylabel('Contagem')
plt.show()
```

Impactos Práticos
Modelagem Estatística:

Assunção de Normalidade: Muitos métodos estatísticos, como ANOVA, testes t e regressões lineares, assumem que os dados são normalmente distribuídos. Se a normalidade é alcançada apenas ao segmentar por categorias, pode ser necessário ajustar modelos separadamente para cada grupo.
Erro Tipo I e Tipo II: Ignorar a não-normalidade pode levar a erros tipo I (falsos positivos) ou tipo II (falsos negativos) ao fazer inferências estatísticas. Isso pode impactar a validade dos resultados.
Interpretação dos Resultados:

Significado das Diferenças: A não-normalidade global pode mascarar diferenças reais entre grupos que são mais visíveis quando analisados separadamente. Isso pode afetar a interpretação de quais fatores são realmente influentes.
Desigualdade nos Grupos: Se os dados forem normalizados dentro de categorias, pode indicar que diferentes grupos têm diferentes padrões ou efeitos que valem a pena ser estudados individualmente.
Decisões de Negócio e Políticas:

Segmentação de Mercado: Em contextos de negócios, entender que diferentes segmentos de clientes ou produtos apresentam características distintas pode ajudar na personalização de estratégias de marketing ou políticas de precificação.
Alocação de Recursos: Se diferentes grupos requerem diferentes abordagens ou recursos, isso pode afetar decisões sobre como alocar melhor os esforços e investimentos.
Abordagens de Modelagem:

Modelos Estratificados: Considerar modelos estatísticos ou de machine learning que tratam as categorias de forma diferenciada, como modelos hierárquicos ou modelos com interação.
Transformações de Dados: Em alguns casos, pode ser possível aplicar transformações aos dados para atingir a normalidade sem segmentar, mas isso pode complicar a interpretação dos resultados.
Exemplos de Abordagem
Ao perceber que a normalidade é atingida apenas ao separar por categorias, algumas abordagens podem ser adotadas:

Análise Separada por Grupo: Conduzir análises estatísticas separadamente para cada grupo categórico.
Incluir Interações em Modelos: Incluir variáveis de interação entre as categorias e outras variáveis no modelo.
Uso de Testes Não Paramétricos: Considerar testes que não assumem normalidade, como o teste de Mann-Whitney ou Kruskal-Wallis, para comparações entre grupos.
Considerações Finais
Reconhecer a influência das variáveis categóricas na distribuição dos dados é crucial para obter insights precisos e evitar conclusões enganosas. Ajustar a análise para considerar essas diferenças pode melhorar significativamente a qualidade dos resultados e a eficácia das decisões baseadas nesses dados.
