```python
# Caminho para salvar os dataframes
pasta = r'C:\Python\df seaborn'

# confere se o caminho já existe
if not os.path.exists(pasta):
    os.makedirs(pasta)
    print(f'Pasta {pasta} criada.')
else:
    print(f'A pasta {pasta} já existe.')
```

```python
tabela = "penguins"
df=sns.load_dataset(tabela)

# Renomear as colunas
df=df.rename(columns={'bill_length_mm': 'compr bico'
                  ,'bill_depth_mm': 'prof bico'
                  ,'flipper_length_mm': 'compr nadadeira'
                  , 'body_mass_g': 'massa'})
df.to_csv(fr'{pasta}\{tabela}.csv')
df.head(2)
```
