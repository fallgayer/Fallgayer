```python

def calculate_statistics(df):
    """
    Calcula estatísticas (assimetria, curtose e erro padrão da média) para cada coluna numérica do DataFrame.

    Parâmetros:
    df (DataFrame): DataFrame contendo os dados.

    Retorna:
    DataFrame: DataFrame contendo as estatísticas calculadas para cada coluna numérica.
    """
    # Seleciona apenas as colunas numéricas
    numeric_columns = df.select_dtypes(include=['number'])

    # Calcula a assimetria para cada coluna numérica
    skewness = numeric_columns.skew()

    # Calcula a curtose para cada coluna numérica
    kurt = numeric_columns.kurtosis()

    # Calcula o erro padrão da média para cada coluna numérica
    std_error_mean = numeric_columns.sem()

    # Cria um DataFrame com as estatísticas calculadas
    statistics_df = pd.DataFrame({
        'Assimetria': skewness,
        'Curtose': kurt,
        'Erro Padrão da Média': std_error_mean
    })

    return statistics_df

# Exemplo de uso:
# Supondo que df seja seu DataFrame
#statistics = calculate_statistics(df)
```

```python
import seaborn as sns
import pandas as pd
from scipy.stats import zscore

# Carregar o dataset "tips"
tips = sns.load_dataset("tips")

# Função para calcular as estatísticas desejadas
def calcular_estatisticas(df):
    # Selecionar apenas as colunas numéricas
    df_numeric = df.select_dtypes(include=['float64', 'int64']).fillna(0)
    
    # Calcular as estatísticas apenas se houver colunas numéricas
    if not df_numeric.empty:
        estatisticas = [
            df_numeric.sum(),
            df_numeric.mean(),
            df_numeric.median(),
            df_numeric.min(),
            df_numeric.max(),
            df_numeric.std(),
            df_numeric.skew(),
            df_numeric.kurtosis(),
            df_numeric.sem()
        ]
        
        # Calculando o Z-score para cada coluna individualmente
        zscore_values = [(df_numeric[col] - df_numeric[col].mean()) / df_numeric[col].std() for col in df_numeric.columns]
        
        # Adicionando os valores do Z-score ao resultado
        estatisticas.append(pd.DataFrame(zscore_values, index=['zscore']).T)
        
        # Converter em DataFrame
        estatisticas_df = pd.concat([pd.DataFrame(estatisticas, index=['sum', 'mean', 'median', 'min', 'max', 'std', 'skew', 'kurtosis', 'sem']), estatisticas[-1]], axis=1)
    else:
        estatisticas_df = pd.DataFrame(columns=['sum', 'mean', 'median', 'min', 'max', 'std', 'skew', 'kurtosis', 'sem', 'zscore'])
    
    return estatisticas_df

# Calcular estatísticas para o DataFrame principal
estatisticas_tips = calcular_estatisticas(tips)

# Mostrar estatísticas para o DataFrame principal
print("Estatísticas para o DataFrame principal:")
print(estatisticas_tips)

# Agrupar por sexo e calcular estatísticas para cada grupo
grupo_sexo = tips.groupby('sex').apply(calcular_estatisticas)

# Resetar índice para melhor visualização
grupo_sexo = grupo_sexo.reset_index(level=1, drop=True)

# Mostrar estatísticas para os grupos
print("\nEstatísticas agrupadas por sexo:")
print(grupo_sexo)
```
